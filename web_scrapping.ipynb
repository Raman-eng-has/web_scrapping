```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept-Encoding": "gzip, deflate, br",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,/;q=0.8",
    "Connection": "keep-alive",
    "Referer": "https://google.com",
}
session=requests.Session()
session.headers.update(headers)
url = 'https://www.niet.co.in/department/artificial-intelligence-machine-language-aiml'
response=session.get(url)
soup=BeautifulSoup(response.text,'lxml')
print(soup)
print('---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
print(soup.title.text)
print('---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
h4=soup.find_all('h4')
print(h4)
print('---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
print(h4[0])
print('---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
h4_update=[]
for i in h4:
  h4_update.append(i.text.strip())
  print(i.text.strip())
print('---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
print(h4_update)
print('---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
print(h4_update[0])
print('---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
for i in soup.find_all('p'):
  print(i.text.strip())
print('---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
for i in soup.find_all('a'):
  print(i.text.strip())
print('---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
for i in soup.find_all('h5'):
  print(i.text.strip())
print('---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
courses=[]
for i in soup.find_all('a',class_='bold'):
  print(i.text.strip())
  courses.append(i.text.strip())
print('---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
df=pd.DataFrame({'courses':courses})
print(df)
print('---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
```
